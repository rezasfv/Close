```python
import openai

# expand query with GPT-3
# use the following prompt for GPT-3 API call (https://beta.openai.com/docs/api-reference/completions/create):
# engine: text-davinci-002 or text-davinci-003
def expand_query(query, num_expansions=20, language="French"):
    openai.api_key = "YOUR TOKEN" # https://platform.openai.com/account/api-keys

    response = openai.Completion.create(
        engine="text-davinci-002",
        # prompt=f"Étendre la requête suivante avec {num_expansions} termes ou expressions liés en {language}: {query}",
        # prompt=f"Act as an user and Expand the following query with {num_expansions} related terms or phrases for information retrieval (search-engine): {query}",
        prompt=f"Expand the following query with {num_expansions} related terms or phrases for information retrieval (search-engine): {query}",
        max_tokens=100,
        n=1,
        stop=None,
        temperature=0.6,
    )

    expanded_terms = response.choices[0].text.strip().split(', ')
    return expanded_terms

title = 'antivirus comparaison'
expanded_terms = expand_query(title)

print(f"Expanded terms for '{title}': {', '.join(expanded_terms)}")

# some examples of expanded terms
# GPT - text-davinci-002 - 1
# 1. Antivirus software
# 2. Antivirus protection
# 3. Best antivirus
# 4. Free antivirus
# 5. Antivirus for Windows
# 6. Antivirus for Mac
# 7. Antivirus for Android
# 8. Mobile antivirus
# 9. Antivirus reviews
# 10. Norton antivirus
# 11. McAfee antivirus
# 12. Kaspersky antivirus
# 13. Bitdefender antivirus
# 14. Avast antivirus

# GPT - text-davinci-003 - 1
# 1. Antivirus Reviews 
# 2. Antivirus Software 
# 3. Antivirus Protection 
# 4. Malware Protection 
# 5. Virus Scanner 
# 6. Online Security 
# 7. Internet Security 
# 8. Firewall Protection 
# 9. Antivirus Download 
# 10. Antivirus Programs 
# 11. Antivirus Solutions 
# 12. Antivirus Protection Comparison 
# 13. Antivirus Comparison Chart

# GPT - text-davinci-002 - 2
# antivirus, 
# anti-malware, 
# anti-spyware, 
# anti-adware, 
# anti-phishing, 
# cyber security, 
# internet security, 
# online security, 
# computer security, 
# information security, 
# data security, 
# mobile security, 
# endpoint security, 
# network security, 
# cloud security, 
# email security, 
# web security, 
# database security
```

```python
import re

# use train.trec file to expand queries
f = open("train.trec", "r")
text = f.read()

# extract queries from train.trec file
pattern = r'<top>\s*<num>(.*?)<\/num>\s*<title>(.*?)<\/title>\s*<\/top>'
matches = re.findall(pattern, text, re.DOTALL)

file1 = open("result.txt", "a")  # append mode

for match in matches:
    num, title = match
    expanded_terms = expand_query(title)
    print(f"Expanded terms for '{title}': {', '.join(expanded_terms)}")
    file1.write(f"{num:} {expanded_terms} \n")

file1.close()
```

```python
# Find the longest common substring in a set of strings to use it as MUST term in Searcher section in Apache Lucene

# Example array of expanded terms
texts = ['antivirus protection', 'antivirus software', 'free antivirus', 'best antivirus', 'antivirus download', 'mobile antivirus', 'antivirus for mac', 'avast antivirus', 'avg antivirus', 'comodo antivirus', 'eset antivirus', 'panda antivirus', 'bitdefender antivirus', 'kaspersky antivirus', 'mcafee antivirus', 'norton antivirus', 'avira antivirus', 'webroot antivirus', 'sophos antivirus']

# Find the shortest text in the array
shortest_text = min(texts, key=len)

# Loop through all possible substrings of the shortest text
longest_substring = ""
for i in range(len(shortest_text)):
    for j in range(i + 1, len(shortest_text) + 1):
        substring = shortest_text[i:j]
        
        # Check if the substring appears in all texts
        if all(substring in text for text in texts):
            if len(substring) > len(longest_substring):
                longest_substring = substring

# longest_substring is the result
print(longest_substring)
```
